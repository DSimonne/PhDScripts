```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import pandas as pd
```

```python
data = np.array([
    [0, 3, 6, -6, -4, -2, -1, -0.5, -0.25, -0.125, 1, 0.5],
    [4.1, 4.239, 5.2, 6.8, 5.47, 4.6, 4.3, 4.2, 3.96, 4.37, 4.35, 4.12]
])
```

```python
df = pd.DataFrame(data.T, columns= ["y", "roi4"])
df = df.sort_values("y")
```

```python
plt.figure(figsize=(16, 9))
plt.plot(df["y"], df["roi4"])
plt.errorbar(df["y"], df["roi4"], np.ones(len(df))*0.1 + 0.01 * df["roi4"]);
plt.tight_layout()
plt.savefig("Images/FindingFocalPlane.png")
```
